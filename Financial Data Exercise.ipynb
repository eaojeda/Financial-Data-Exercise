```python
from sklearn import preprocessing
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
financial_data = pd.read_csv('financial_data.csv')
```

```python
financial_data
```

```python
month = financial_data['Month']
revenue = financial_data['Revenue']
expenses = financial_data['Expenses']
```

```python
fig,ax = plt.subplots()
ax.invert_yaxis()
plt.plot(month, revenue)
plt.xlabel('Month')
plt.ylabel('Amount ($)')
plt.title('Revenue')
plt.show()
plt.clf()
```

```python
fig,ax = plt.subplots()
plt.plot(month, expenses)
plt.xlabel('Month')
plt.ylabel('Amount ($)')
plt.title('Expenses')
plt.show()
plt.clf()
```
